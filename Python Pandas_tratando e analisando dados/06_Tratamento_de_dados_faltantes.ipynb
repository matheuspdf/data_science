# 06. Tratamento de dados faltantes

### Nesta aula:

* Tratamento de dados faltantes
* Como identificar valores nulos (missing values)
    * O método isnull() indica se os valores são nulos
    * O método notnull() retorna o contrário do método isnull()
    * O método info() também é uma forma de se verificar a presença de valores 
    
* Como remover valores nulos com o método dropna()

* Tratamento condicional
* Inversão de valores booleanos com ~
* Como substituir os missing values com o método fillna()
* Métodos de interpolação: ffill, bfill e mean()

```python
import pandas as pd
dados = pd.read_csv('aluguel.csv', sep = ';')
dados
```

```python
dados.isnull() # cria um DF booleano / notnull() -> retorna o contrário
```

```python
dados.info() # retorna informações do DF
```

```python
dados[dados['Valor'].isnull()] # retorna os registro que possuem Nulo na coluna específica
```

```python
A = dados.shape[0]
dados.dropna(subset = ['Valor'], inplace = True)
B = dados.shape[0]
A - B
```

```python
dados[dados['Valor'].isnull()] # retorna um DT vazio
```

```python
#quantos condomínios com o valor Null
dados[dados['Condominio'].isnull()].shape[0]
```

```python
selecao = (dados['Tipo'] == 'Apartamento') & (dados['Condominio'].isnull())
```

```python
A = dados.shape[0]
dados = dados[~selecao]
B = dados.shape[0]
A - B
```

```python
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados = dados.fillna({'Condominio': 0, 'IPTU': 0})
dados[dados['Condominio'].isnull()].shape[0]
```

```python
dados[dados['IPTU'].isnull()].shape[0]
dados.info()
```

```python
dados.to_csv('aluguel_sem_null.csv', sep = ';', index = False)
```

# (Extra) Método de interpolação



```python
import pandas as pd

data = [0.5, None, None, 0.52, 0.54, None, 
        None, 0.56, 0.59, None, None, 0.6, None, 0.9]

s = pd.Series(data)
s
```

```python
s.fillna(0) #preenche com o valor do parênteses
```

```python
s.fillna(method = 'ffill')# pega o valor válido anterior e substitui
```

```python
s.fillna(method = 'bfill') # preenche com o valor válido, começando debaixo
```

```python
s.fillna(s.mean()) # média dos valores não nulos como preenchimento
```

```python
s1 = s.fillna(method = 'ffill', limit = 1) # preenche com um valor válido só uma vez
```

```python
s1.fillna(method = 'bfill', limit = 1) # preenche com um valor válido só uma vez, começando debaixo
```
