Nesta aula, vimos como:

* Ordenar os dados de um dataframe (métodos sort_values() e sort_index())

## Imóveis residenciais 

```python
import pandas as pd
dados = pd.read_csv('aluguel.csv', sep = ';')
dados.head(10)
```

```python
list(dados['Tipo'].drop_duplicates())
```

```python
residencial= ['Quitinete','Casa','Apartamento','Casa de Condomínio','Casa de Vila']

residencial
```

```python
dados.head(10)
```

```python
selecao = dados["Tipo"].isin(residencial)
selecao
```

```python
dados_residencial = dados[selecao]
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
dados_residencial.shape[0]
```

```python
dados.shape[0]
```

```python
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial
```

### Exportando a Base de Dados

```python
dados_residencial.to_csv('aluguel_residencial.csv', sep=';', index = False)
```

```python
dados_residencial_2 = pd.read_csv('aluguel_residencial.csv', sep=';')
dados_residencial_2
```
